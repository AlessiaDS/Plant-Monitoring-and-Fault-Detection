**Confusion matrix**

matrix.png

```python
import numpy as np

confusion_matrix = np.array([[2542, 2, 1, 15],
                             [7, 1231, 40, 2],
                             [1, 45, 1230, 4],
                             [36, 6, 3, 1235]])
```

Metrics for class A

```python
tpA = confusion_matrix[0,0]
tnA = confusion_matrix[1,1] + confusion_matrix[1,2] + confusion_matrix[1,3] + confusion_matrix[2,1] + \
      confusion_matrix[2,2]+ confusion_matrix[2,3] + confusion_matrix[3,1] + confusion_matrix[3,2] +  \
      confusion_matrix[3,3]
fpA = confusion_matrix[1,0] + confusion_matrix[2,0] + confusion_matrix[3,0]
fnA = confusion_matrix[0,1] + confusion_matrix[0,2] + confusion_matrix[0,3]
```

Metrics for class B

```python
tpB = confusion_matrix[1,1]
tnB = confusion_matrix[0,0] + confusion_matrix[0,2] + confusion_matrix[0,3] + confusion_matrix[2,0] + \
      confusion_matrix[2,2]+ confusion_matrix[2,3] + confusion_matrix[3,0] + confusion_matrix[3,2] +  \
      confusion_matrix[3,3]
fpB = confusion_matrix[0,1] + confusion_matrix[2,1] + confusion_matrix[3,1]
fnB = confusion_matrix[1,0] + confusion_matrix[1,2] + confusion_matrix[1,3]
```

Metrics for class C

```python
tpC = confusion_matrix[2,2]
tnC = confusion_matrix[0,0] + confusion_matrix[0,1] + confusion_matrix[0,3] + confusion_matrix[1,0] + \
      confusion_matrix[1,1]+ confusion_matrix[1,3] + confusion_matrix[3,0] + confusion_matrix[3,1] +  \
      confusion_matrix[3,3]
fpC = confusion_matrix[0,2] + confusion_matrix[1,2] + confusion_matrix[3,2]
fnC = confusion_matrix[2,0] + confusion_matrix[2,1] + confusion_matrix[2,3]
```

Metrics for class D

```python
tpD = confusion_matrix[3,3]
tnD = confusion_matrix[0,0] + confusion_matrix[0,1] + confusion_matrix[0,2] + confusion_matrix[1,0] + \
      confusion_matrix[1,1]+ confusion_matrix[1,2] + confusion_matrix[3,0] + confusion_matrix[3,1] +  \
      confusion_matrix[3,2]
fpD = confusion_matrix[0,3] + confusion_matrix[1,3] + confusion_matrix[2,3]
fnD = confusion_matrix[3,0] + confusion_matrix[3,1] + confusion_matrix[3,2]
```

Metrics for multi classes

```python
tp = [tpA, tpB, tpC, tpD]
tn = [tnA, tnB, tnC, tnD]
fp = [fpA, fpB, fpC, fpD]
fn = [fnA, fnB, fnC, fnD]
```

Average accuracy

acc.jpeg

```python
acc = 0
for i in range(4):
  acc = acc + 1/4*((tp[i]+tn[i])/(tp[i]+tn[i]+ fp[i]+fn[i]))
acc_perc = acc * 100
print(acc_perc)
```

Average precision (macro-averaging)

ppv.jpeg

```python
ppv_M=0
for i in range(4):
  ppv_M = ppv_M + 1/4*(tp[i]/(tp[i]+ fp[i]))
ppv_M_perc = ppv_M * 100
print(ppv_M_perc)
```

Average precision (micro-averaging)

ppv2.jpeg

```python
ppv_mu=0
tpi= 0
tfpi= 0
for i in range(4):
  tpi=tpi +tp[i]
  tfpi= tfpi + tp[i]+ fp[i]
ppv_mu= tpi/tfpi
ppv_mu_perc = ppv_mu *100
print(ppv_mu_perc)
```

Average recall (macro-averaging)

tpr.jpeg

```python
tpr_M=0
for i in range(4):
  tpr_M = tpr_M + 1/4*(tp[i]/(tp[i]+fn[i]))
tpr_M_perc = tpr_M * 100
print(tpr_M_perc)
```

Average recall (micro-averaging)

tpr2.jpeg

```python
tpr_mu=0
tpi= 0
tpfni= 0
for i in range(4):
  tpi=tpi +tp[i]
  tpfni= tpfni + tp[i]+ fn[i]
tpr_mu= tpi/tpfni
tpr_mu_perc = tpr_mu *100
print(tpr_mu_perc)
```

Average F1 score (macro-averaging)

f1.jpeg

```python
F1_M = (2 * ppv_M * tpr_M)/(ppv_M + tpr_M)
F1_M_perc = F1_M * 100
print(F1_M_perc)
```

Average F1 score (micro-averaging)

f12.jpeg

```python
F1_mu = (2 * ppv_mu * tpr_mu)/(ppv_mu + tpr_mu)
F1_mu_perc = F1_mu * 100
print(F1_mu_perc)
```

Average specificity (macro-averaging)

tnr.jpeg

```python
tnr_M=0
for i in range(4):
  tnr_M = tnr_M + 1/4*(tn[i]/(tn[i]+fp[i]))
tnr_M_perc = tnr_M * 100
print(tnr_M_perc)
```

Average specificity (micro-averaging)

tnr2.jpeg

```python
tnr_mu=0
tni= 0
tnfpi= 0
for i in range(4):
  tni=tni +tn[i]
  tnfpi= tnfpi + tn[i]+ fp[i]
tnr_mu= tni/tnfpi
tnr_mu_perc = tnr_mu *100
print(tnr_mu_perc)
```
